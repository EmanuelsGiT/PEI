```python

# Read the first dataset
df1 = pd.read_csv('../measurements_final/measurements_haskell.csv')

# Read the second dataset
df2 = pd.read_csv('../measurements_python.csv')

# Read the third dataset
df3 = pd.read_csv('../measurements_final/measurements_java.csv')

df4 = pd.read_csv('../measurements_final/measurements_rust.csv')

df5 = pd.read_csv('../measurements_final/measurements_ruby.csv')

df6 = pd.read_csv('../measurements_final/measurements_c.csv')

# Concatenate the datasets vertically
combined_df = pd.concat([df1, df2, df3, df4, df5, df6], ignore_index=True)
combined_df = combined_df.drop(["DRAM",'GPU'], axis=1)

dfd11 = combined_df.drop(["Language",'Program'], axis=1)

print(dfd11.head())


correlation_table = dfd11.corr()
correlation_table





# Display the combined dataset
```

```python
df = df.drop(columns=['GPU', 'DRAM', 'Language'])

df.shape
```

```python
def remove_outliers(group):
    sorted_group = group.sort_values(by='Package')
    return sorted_group.iloc[2:-2]

# Group by 'Program' and 'PowerLimit' and apply the function to remove outliers
df = df.groupby(['Program', 'PowerLimit']).apply(remove_outliers).reset_index(drop=True)

# Display the cleaned DataFrame
print(df)
```

```python
df_mean = df.groupby(['Program', 'PowerLimit']).mean().reset_index()

df_mean.shape
```

```python
# Filter for "Haskell" language
filtered_haskell = filtered_df[filtered_df["Language"] == "Haskell"]

# Include PowerLimit values -1, 2, and 4
filtered_haskell = filtered_haskell[filtered_haskell['PowerLimit'].isin([-1, 2, 4])]

# Calculate the mean values for "Package" and "Core" columns
mean_package = filtered_haskell.groupby(['Program', 'PowerLimit'])['Package'].mean().unstack()
mean_core = filtered_haskell.groupby(['Program', 'PowerLimit'])['Core'].mean().unstack()

# Create a bar chart for "Package" and "Core" means
fig, ax = plt.subplots(figsize=(23, 10))

# Plot the "Package" values
mean_package.plot(kind='bar', ax=ax, width=0.4, label=['Package -1', 'Package 2', 'Package 4'])

# Plot the "Core" values with shifted positions for bars
mean_core.plot(kind='bar', color=['cyan', 'orange', 'purple'], ax=ax, width=0.4, label=['Core -1', 'Core 2', 'Core 4'])

# Calculate the mean values for "Time"
mean_time = filtered_haskell.groupby(['Program', 'PowerLimit'])['Time'].mean().unstack()

# Create a second y-axis for the "Time" values
plot2 = ax.twinx()
mean_time.plot(kind='line', color=['green', 'pink', 'yellow'], ax=plot2, label=['Time -1', 'Time 2', 'Time 4'])

# Set the title and labels
ax.set_title('Package and Core Mean Values for Each Program (PowerLimits -1, 2, 4)')
ax.set_xlabel('Program')
ax.set_ylabel('Package and Core Values')
plot2.set_ylabel('Time (segundos)')

# Add legends
ax.legend(loc='upper left')
plot2.legend(loc='upper right')

# Show the plot
plt.savefig('graph_haskell.png', dpi=300, bbox_inches='tight')
plt.show()
```
